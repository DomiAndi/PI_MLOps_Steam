**Proceso de ETL: User_Items**

**Cargamos las librerias necesarias**

```python
import pandas as pd
from pandas import json_normalize
import numpy as np
import json
import gzip
import warnings
warnings.filterwarnings("ignore")
import ast
```

**Cargamos el archivo json y pasamos los datos a un dataframe**

```python
#Leer el archivo JSON comprimido
def leer_jsoncomprimido():
    with gzip.open('../Archivos Json/users_items.json.gz','rt',encoding='utf-8')as file:
        return[ast.literal_eval(line.strip())
               for line in file]

filas_items = leer_jsoncomprimido()

#convertirlo a un dataframe
df_items=pd.DataFrame(filas_items)
```

Como se observo en la linea de codigo anterior la variable items contiene varios diccionarios dentro de un mismo registro por lo que se procede a desanidar la misma para su posterior analisis

```python
df_items= df_items.explode("items").reset_index(drop=True)
df_items
```

```python
df_items_normalizado = pd.json_normalize(df_items["items"])
df_items_normalizado
```

```python
df_items_final = pd.concat([df_items,df_items_normalizado], axis=1)
df_items_final
```

Eliminamos la columna items dado que ya no es necesaria

```python
df_items_final = df_items_final.drop(columns=["items"])
df_items_final
```

**Se procede a eliminar columnas que no son relevantes para nuestro proposito como "user_url" y "playtime_2weeks"**

```python
df_items_final = df_items_final.drop(columns=["user_url","playtime_2weeks","steam_id"])
df_items_final
```

```python
df_items_final.info()
```

**Corroboramos sobre la variable "item_name" que tengamos registros que se puedan usar para el modelado**

```python
df_items_final["item_name"].unique()
```

```python
patron_no_deseado = r'[^a-zA-Z0-9\s]'  
df_filtrado = df_items_final[df_items_final["item_name"].str.contains(patron_no_deseado) == False]
df_filtrado
```

**Se verifica si tenemos registros nulos**

```python
df_filtrado.isnull().sum()
```

**Se verifica si tenemos registros duplicados**

```python
duplicados = df_filtrado.duplicated().sum()
duplicados
```

```python
duplicados_mask = df_filtrado.duplicated(keep=False)
filas_duplicadas = df_filtrado[duplicados_mask]
filas_duplicadas
```

En este caso verificamos que las filas que aparecen como duplicadas corresponden al mismo usuario por lo que las variables: user_id, items_count, steam_id son las mismas pero varia el juego al cual se esta refiriendo por lo que se decide conservarlas

**Carga del conjunto de datos**

```python
archivo_limpio = "user_items_limpio.csv"
df_filtrado.to_csv(archivo_limpio, index=False, encoding="utf-8")
print(f"Se guardó el archivo {archivo_limpio}")
```
